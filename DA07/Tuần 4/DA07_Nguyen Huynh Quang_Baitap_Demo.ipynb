```python
#Tạo phương trình
a = int(input("Nhập tham số a: "))
b = int(input("Nhập tham số b: "))
x = np.linspace(-10, 10, 100)
y = a*x + b
#Tạo hệ trục toạ độ 
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# Vẽ đồ thị phương trình
plt.title("Phương trình y = ax + b")
plt.plot(x,y, 'r', label = 'y = ax + b')
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

```python
with open('excel_fpt.csv', 'r', encoding='utf-8', newline = '') as f:
    data_reader = csv.reader(f)
    date = []
    ds_open = []
    ds_close = []
    ds_high = []
    ds_low = []
    volume = []
    for i, row in enumerate(data_reader):
        date.append(row[1])
        ds_open.append(row[7])
        ds_close.append(row[10])
        ds_high.append(row[8])
        ds_low.append(row[9])
        volume.append(row[11])
    # Remove title
    date.pop(0)
    ds_open.pop(0)
    ds_close.pop(0)
    ds_high.pop(0)
    ds_low.pop(0)
    volume.pop(0)
    #Convert from string to integer
    ds_open = [float(x) for x in ds_open]
    ds_close = [float(y) for y in ds_close]
    ds_high = [float(a) for a in ds_high]
    ds_low = [float(b) for b in ds_low]
    volume = [int(c) for c in volume]
    date = [int(d) for d in date]
    
```

```python
#Bài 3b
x = np.linspace(-5,5,100)
a = int(input("Nhập tham số a"))
b = int(input("Nhập tham số b"))
c = int(input("Nhập tham số c"))
y = a*(x**2) + b*x + c
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
# Vẽ hệ trục toạ độ
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# Vẽ đồ thị phương trình
plt.title("Phương trình a*x^2 + b*x + c")
plt.plot(x,y,'r', label ="y = a*x^2 + b*x + c")
plt.legend()
plt.grid()
plt.show()
```
