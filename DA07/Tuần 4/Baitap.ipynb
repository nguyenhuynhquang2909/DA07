```python
#Tần suất giứoi tính
with open('supermarket_sales_vn.csv', 'r', encoding = 'utf-8', newline = '') as f:
    f_csv = csv.reader(f)
    customer = []
    for i, line in enumerate(f_csv):
        customer.append(line[3])
    customer.pop(0)
    count_male = 0
    count_female = 0
    for gender in customer:
        if gender == "Nam": count_male += 1
        if gender == "Nữ": count_female += 1
    x = np.random.randn(count_male)
    y = np.random.randn(count_female)
    plt.hist(x, facecolor = 'red', alpha = 0.75, density = True, label = "Nam")
    plt.hist(y, facecolor = 'blue', alpha = 0.75, density = True, label = "Nữ")
    plt.legend()
    plt.show()
```
